```python
def get_orthogonal_patches(x):
    dims = x.shape # Get shape of X
    axial = x[dims[0]//2,:,:]. squeeze()
    coronal = x[:,:,dims[2]//2].squeeze()
    sagittal= x[:,dims[1]//2,:].squeeze()
    print('orthogonal shape',dims[0]//2)
    return axial, coronal, sagittal
```
